## DBSCAN from the scikit-learn

Here I am using the DBSCAN to find clusters in my super-resolution dataset.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.cluster import dbscan
```

#### Loading the data
The dataset consists of many columns. Each columns contains information regarding each spot image detected in the dataset. The 4th and the 5th column contain the X and Y location of the spots that I will use.

```python
data = np.genfromtxt('play1.txt',skip_header=1,usecols=(3,4))
print ("Number of spots in the dataset = {}".format(data.shape[0]))
```

#### Running DBSCAN
Here I am setting a distance of 0.5 to consider points to be linked to each other, and atleast 50 points to be linked to be considered a cluster.

```python
cs,labels = dbscan(data,eps=0.5,min_samples=50)
```

Separating the clusters and the noise

```python
ctrue = data[labels!=-1,:]
ctrue_labels = labels[labels!=-1]
cfalse = data[labels==-1,:]
cfalse_labels = labels[labels==-1]
```

#### Plotting the results of DBSCAN

```python

plt.figure(1,figsize=(9,4))
plt.subplot(1,2,1)
plt.scatter(ctrue[:,0],ctrue[:,1],15,ctrue_labels)
plt.plot(cfalse[:,0],cfalse[:,1],'.y',ms=2)
plt.xlim([60,90])
plt.ylim([95,125])
plt.title("Clusters from DBSCAN")

plt.subplot(1,2,2)
plt.plot(data[:,0],data[:,1],'.b',ms=5)
plt.xlim([60,90])
plt.ylim([95,125])
plt.title("Original data")

plt.show()
```
